## Amostragem

Separe os trs ltimos meses como safras de validao *out of time* (oot).

Variveis:<br>
Considere que a varivel ```data_ref``` no  uma varivel explicativa,  somente uma varivel indicadora da safra, e no deve ser utilizada na modelagem. A varivei ```index```  um identificador do cliente, e tambm no deve ser utilizada como covarivel (varivel explicativa). As restantes podem ser utilizadas para prever a inadimplncia, incluindo a renda.


````python
# Amostragem (DEV / TEST / OOT)

# Garantir tipo datetime
df['data_ref'] = pd.to_datetime(df['data_ref'])

# Criar coluna ano-ms para definir safras
df['ano_mes'] = df['data_ref'].dt.to_period('M')

# Ordenar e separar ltimas 3 safras como OOT
meses_ordenados = np.sort(df['ano_mes'].unique())
oot_safras = meses_ordenados[-3:]  # ltimas 3 safras para validao out-of-time

mask_oot = df['ano_mes'].isin(oot_safras)
mask_dev = ~mask_oot

dev = df.loc[mask_dev].copy()
oot = df.loc[mask_oot].copy()

print('Safras OOT:', list(map(str, oot_safras)))
print('Tamanhos -> DEV:', dev.shape, '| OOT:', oot.shape)

# Separar variveis explicativas e alvo
alvo = 'mau'
to_drop = ['data_ref', 'ano_mes', 'index', alvo]

X_dev = dev.drop(columns=to_drop, errors='ignore')
y_dev = dev[alvo].astype(int)

X_oot = oot.drop(columns=to_drop, errors='ignore')
y_oot = oot[alvo].astype(int)

# Split do DEV em treino e teste (80/20)
X_train, X_test, y_train, y_test = train_test_split(
    X_dev, y_dev, test_size=0.2, random_state=42, stratify=y_dev
)

# Detectar tipos de variveis
num_cols = X_dev.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = [c for c in X_dev.columns if c not in num_cols]

print(f'Numricas: {len(num_cols)} | Categricas: {len(cat_cols)}')
print(f'Train: {X_train.shape} | Test: {X_test.shape} | OOT: {X_oot.shape}')
````

## Descritiva bivariada

Faa uma anlise descritiva bivariada de cada varivel

````python
# EDA bivariada: taxa de evento por categoria / por faixas (qcut) para numricas.
def taxa_evento_por_cat(df_, col, y='y'):
    tab = (df_.groupby(col)[y]
           .agg(['count','mean'])
           .rename(columns={'count':'n','mean':'taxa_evento'})
           .sort_values('taxa_evento', ascending=False))
    return tab

# Preparar base de dev com alvo renomeado para 'y' para facilitar
tmp = X_dev.copy()
tmp['y'] = y_dev.values

# Categricas
for c in cat_cols[:10]:
    print(f'\nTaxa de evento por {c}:')
    display(taxa_evento_por_cat(tmp, c, y='y').head(15))

# Numricas -> qcut em 10 faixas
for c in num_cols[:10]:
    if is_numeric_dtype(tmp[c]):
        try:
            tmp[f'{c}_bin'] = pd.qcut(tmp[c], q=10, duplicates='drop')
            tab = taxa_evento_por_cat(tmp, f'{c}_bin', y='y')
            print(f'\nTaxa de evento por faixas de {c}:')
            display(tab.head(15))
        except Exception as e:
            print(f'[Aviso] No consegui binar {c}:', e)
````

## Desenvolvimento do modelo

Desenvolva um modelo de *credit scoring* atravs de uma regresso logstica.

- Trate valores missings e outliers
- Trate 'zeros estruturais'
- Faa agrupamentos de categorias conforme vimos em aula
- Proponha uma equao preditiva para 'mau'
- Caso hajam categorias no significantes, justifique

````python
# Diviso entre treino e validao dentro do DEV
X_tr, X_va, y_tr, y_va = train_test_split(
    X_dev, y_dev, test_size=0.25, random_state=42, stratify=y_dev
)


# Pr-processamento
num_tf = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),     # substitui NaN pela mediana
    ('scaler', StandardScaler())
])

cat_tf = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),  # substitui NaN pela categoria mais comum
    ('ohe', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

pre = ColumnTransformer(
    transformers=[
        ('num', num_tf, num_cols),
        ('cat', cat_tf, cat_cols)
    ],
    remainder='drop'
)


# Modelo
logreg = LogisticRegression(max_iter=200, class_weight='balanced')

pipe = Pipeline(steps=[
    ('pre', pre),
    ('clf', logreg)
])

pipe.fit(X_tr, y_tr)


# Avaliao
proba_tr = pipe.predict_proba(X_tr)[:, 1]
proba_va = pipe.predict_proba(X_va)[:, 1]

# AUC (com formatao correta)
auc_tr = roc_auc_score(y_tr, proba_tr)
auc_va = roc_auc_score(y_va, proba_va)
print(f'AUC (treino): {auc_tr:.3f}')
print(f'AUC (val)   : {auc_va:.3f}')


# Cross-validation (5 folds)
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
auc_cv = cross_val_score(pipe, X_dev, y_dev, cv=cv, scoring='roc_auc')
print('AUC ROC (CV 5-fold):', np.round(auc_cv, 3))
print(f'Mdia  DP: {auc_cv.mean():.3f}  {auc_cv.std():.3f}')


# Probabilidades no OOT
proba_oot = pipe.predict_proba(X_oot)[:, 1]
# Garantir train/test para o DEV
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_dev, y_dev, test_size=0.2, random_state=42, stratify=y_dev)
````

### Remoo de outliers

Como identificar outlier? Substituir o outlier por algum valor? Remover a linha?

````python
def winsorize_df(df, cols, q_low=0.01, q_high=0.99):
    df_w = df.copy()
    for c in cols:
        ql, qh = df[c].quantile(q_low), df[c].quantile(q_high)
        df_w[c] = df[c].clip(ql, qh)
    return df_w
````

### Seleo de variveis

Qual tipo de tcnica? Boruta? Feature importance? 

````python
# 1. Cpia do conjunto de treino
X_train_prep = X_train.copy()
y_train_enc = y_train.copy()

# 2. Tratamento de valores ausentes
# - Preenche variveis numricas com a mdia
# - Preenche variveis categricas com "Desconhecido"
for col in X_train_prep.columns:
    if X_train_prep[col].dtype == 'object':
        X_train_prep[col] = X_train_prep[col].fillna('Desconhecido')
    else:
        X_train_prep[col] = X_train_prep[col].fillna(X_train_prep[col].mean())

# 3. Codificao das variveis categricas em dummies
X_train_enc = pd.get_dummies(X_train_prep, drop_first=True)

# 4. Treinamento do modelo base (Random Forest)
rf = RandomForestClassifier(
    n_estimators=200,
    random_state=42,
    n_jobs=-1
)
rf.fit(X_train_enc, y_train_enc)

# 5. Obteno das importncias das variveis
importancias = pd.DataFrame({
    'Varivel': X_train_enc.columns,
    'Importncia': rf.feature_importances_
}).sort_values('Importncia', ascending=False)

# 6. Exibir top 10 variveis mais relevantes
display(importancias.head(10))

# 7. Interpretao
print("\nAs variveis acima apresentaram maior importncia no modelo Random Forest,")
print("indicando forte poder preditivo para o risco de crdito.")
````

### Pipeline 

Crie um pipeline contendo essas funes.

preprocessamento()
- substituicao de nulos
- remoo outliers
- PCA
- Criao de dummy de pelo menos 1 varivel (posse_de_veiculo)

````python
num_tf = Pipeline(steps=[
    ('imputer', __import__('sklearn.impute').impute.SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])
cat_tf = Pipeline(steps=[
    ('imputer', __import__('sklearn.impute').impute.SimpleImputer(strategy='most_frequent')),
    ('ohe', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
]) if len(cat_cols) > 0 else 'drop'

pre = ColumnTransformer(
    transformers=[('num', num_tf, num_cols),
                  ('cat', cat_tf, cat_cols)],
    remainder='drop'
)

clf = Pipeline(steps=[
    ('pre', pre),
    ('lr', LogisticRegression(max_iter=200, class_weight='balanced'))
])

clf.fit(X_train, y_train)

def ks_statistic(y_true, y_score):
    fpr, tpr, _ = roc_curve(y_true, y_score)
    return (tpr - fpr).max()

def eval_block(X_, y_, tag):
    proba = clf.predict_proba(X_)[:, 1]
    pred  = (proba >= 0.5).astype(int)
    auc = roc_auc_score(y_, proba)
    ks  = ks_statistic(y_, proba)
    ap  = average_precision_score(y_, proba)
    acc = accuracy_score(y_, pred)
    f1  = f1_score(y_, pred)
    print(f"[{'{'}tag{'}'}] AUC={auc:.4f} | KS={ks:.3f} | AP={ap:.4f} | ACC={acc:.4f} | F1={f1:.4f}")
    return {'AUC': auc, 'KS': ks, 'AP': ap, 'ACC': acc, 'F1': f1}

metrics = {}
metrics['DEV'] = eval_block(X_train, y_train, 'DEV')
metrics['TEST'] = eval_block(X_test, y_test, 'TEST')
if 'X_oot' in globals() and X_oot is not None:
    metrics['OOT'] = eval_block(X_oot, y_oot, 'OOT')
````
